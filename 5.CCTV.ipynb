```python
# CCTV 각 구 별로 2013년도 이전 대비 - 이후 증가율 : column = 최근 증가율 

CCTV_Seoul['최근증가율'] = (CCTV_Seoul['2016년'] + CCTV_Seoul['2015년'] + CCTV_Seoul['2014년']) / CCTV_Seoul['2013년도 이전'] * 100
CCTV_Seoul.head()
```

```python
# 증가율 = (비교대상-기준)/기준 * 100 

CCTV_Seoul['2014증가율'] = (CCTV_Seoul['2013년도 이전'] - CCTV_Seoul['2014년']) / CCTV_Seoul['2013년도 이전'] * 100
CCTV_Seoul['2015증가율'] = (CCTV_Seoul['2013년도 이전'] - CCTV_Seoul['2015년']) / CCTV_Seoul['2013년도 이전'] * 100
CCTV_Seoul['2016증가율'] = (CCTV_Seoul['2013년도 이전'] - CCTV_Seoul['2016년']) / CCTV_Seoul['2013년도 이전'] * 100

CCTV_Seoul.head()
```

```python
plt.figure(figsize=(6,6))

# scatter = 점으로 그리기 
# s = 점의 크기
plt.scatter(data_result['인구수'], data_result['소계'], s=50)

plt.xlabel('인구수')
plt.ylabel('CCTV') #cctv총개수
plt.grid()
plt.show()
```

```python
# 선형회귀 ; linear Regression 
# 1 : 1차 함수 : ax+b
# x = 인구수
# y = 소개

fp1 = np.polyfit(data_result['인구수'], data_result['소계'], 1)
# polyfit이 직선을 만들기 위한 계수(기울기와 y-bias값)를 찾아 주고,
# poly1d가 직선의 계수를 이용해서 1차함수를 만들어 주는 것
# linspace는 그림을 그리기 위해 x축 데이터를 만든 것

# f1은 polyfit 명령 마치고 , 함수
```

```python
fp1
# a, b 깂
```

```python
type(fp1)
```

```python
# y 축 data
f1 = np.poly1d(fp1)
# poly1d이용 polynomial class

# x 축 data
fx = np.linspace(100000, 700000, 100)
```

```python
# 인구수가 310,000 일때, CCTV 수는 1100대가 적당 하다 

plt.figure(figsize=(8,8))
plt.scatter(data_result['인구수'], data_result['소계'], s=50)
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')
#            f1은 polyfit 명령 마치고 , 함수 되었음 
plt.xlabel('인구수')
plt.ylabel('CCTV') #cctv총개수
plt.grid()
plt.show()
```

```python
# (설들력 있는 자료 만들기)
# 오차를 계산할수 있는 코드 작성 후  --> 오차가 큰 순으로 데이터 정렬

# abs = absolute value (- --> +)

data_result['오차'] = np.abs(data_result['소계'] - f1(data_result['인구수']))
df_sort = data_result.sort_values(by='오차', ascending=False)
df_sort.head()
```

```python
# https://matplotlib.org/stable/api/_as_gen/matplotlib.pyplot.text.html?highlight=pyplot%20text#matplotlib.pyplot.text

# plot 크기 설정
plt.figure(figsize=(14,10))

# 산점도 scatter 
plt.scatter(data_result['인구수'], data_result['소계'], c=data_result['오차'], s=50)

# 회귀선
plt.plot(fx, f1(fx), ls='dashed', lw=3, color='g')

# 오차가 가장 큰 10개 지역구 이름 출력
for n in range(10): 
    plt.text(df_sort['인구수'][n] * 1.02, 
             # x , 강남구(570500), 송파구(667483), 양천구
             df_sort['소계'][n] * 0.98, 
             # y , 강남구(2780), 송파구(618), 양천구
             df_sort.index[n], 
             # 강남구, 송파구, 양천구 
             fontsize=15)
# 1.02, 0.98이라는 숫자가 없으면, 
# 글자가 그림을 가려서 약간 옆으로 이동하라

plt.xlabel('인구수')
plt.ylabel('인구당비율')
# 맨 오른쪽 생상바
plt.colorbar()
plt.grid()
plt.show()
```
