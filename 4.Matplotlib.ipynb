```python
# sub plot (하나면에 4개 그리기)

x = np.arange(0, 10, 0.1)
y1 = 0.3*(x-5)** + 1
y2 = -1.5*x + 3
y3 = np.sin(x)**2
y4 = 10*np.exp(-x) + 1
```

```python
# 좌 - 우 - 위 - 아래 

plt.subplot(2,2,1)  #subplot(Figure 안의 행, 열,1st subplot)
plt.plot(x, y1)

plt.subplot(2,2,2)
plt.plot(x, y2)

plt.subplot(2,2,3)
plt.plot(x, y3)

plt.subplot(2,2,4)
plt.plot(x, y4)

plt.show()
```

```python
plt.subplot(3,3,1)  #subplot(**Figure 안의** 행, 열,1st subplot)
plt.plot(x, y1)

plt.subplot(3,3,2)
plt.plot(x, y2)

plt.subplot(3,3,3)
plt.plot(x, y3)

plt.subplot(3,3,4)
plt.plot(x, y4)

plt.show()
```

```python
# Decoration 
#           순서 상관 x 
# https://matplotlib.org/stable/gallery/style_sheets/style_sheets_reference.html 

# 1)color 
# m = magenta , y = yellow, k = kaki

plt.plot(x, y1, 'm', x, y2,'y')
plt.show()
```

```python
# 2)line style

plt.plot(x, y1, '-', x, y2,'--')
plt.show()
```

```python
# 3)marker           
#           ^(triangle up) s(squrare) d(diamond) 
# https://matplotlib.org/stable/api/markers_api.html

x = np.arange(0, 5, 1)
y1 = x
y2 = x + 1

plt.plot(x, y1, '*', x, y2,'1')
plt.show()
```

```python
# 순서 : color, line style, marker    #d:b 

plt.plot(x, y1, 'm--*', x, y2,'s-g') 
plt.show()
```

```python
x = np.arange(-3, 5.5, 0.2)
y = 2*x**3
```

```python
plt.plot(x,y)

# lable 지정
plt.xlabel('X-AXIS')
plt.ylabel('Y-AXIS')

# title 지정
plt.title('Graph title')

# grid 지정
plt.grid(True)
plt.show()
```

```python
# Legend = 항목 설명

x = np.arange(0, 5, 1)
y1 = x
y2 = x + 1
```

```python
plt.plot(x, y1, 'm--*', x, y2,'s-g') 
plt.legend(['data1', 'data2'])
plt.show()
```
