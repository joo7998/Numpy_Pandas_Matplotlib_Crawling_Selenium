```python
# pip install bs4

import urllib.request
import urllib.parse
from bs4 import BeautifulSoup
```

```python
header = {'User-Agent': 'Mozilla/5.0'}
# HTTP Error 406: Not Acceptable  = Crawling 을 막아놓은 싸이트들 
# User-agent 헤더 = 나는 로봇 x , User info(0) 사람 

url = 'https://www.melon.com/chart/'
req = urllib.request.Request(url, headers = header)
html = urllib.request.urlopen(req).read()
```

```python
# 확인 해보기 
print(html)
```

```python
soup = BeautifulSoup(html, 'html.parser')
```

```python
# soup
print(soup.title.text)
```

```python
titles = soup.find_all("div", {"class": "ellipsis rank01"})
singers = soup.find_all("div", {"class": "ellipsis rank02"})
albums = soup.find_all("div",{"class": "ellipsis rank03"})
 
title = []
singer = []
album = []

for t in titles:
    title.append(t.find('a').text)
 
for s in singers:
    singer.append(s.find('a').text)

for a in albums:
    album.append(a.find('a').text)

rank = 100
for i in range(rank): 
    print('%3d위: %s [ %s ] => %s'%(i+1, title[i], singer[i], album[i]))
```
